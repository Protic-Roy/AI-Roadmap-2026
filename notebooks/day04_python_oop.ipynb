```python
class Book:

    def __init__(self,title,author):
        self.title = title
        self.author = author

book1 = Book("Python Basic","Roy")
book2 = Book("AI Engineering","Andrew")

print(f"{book1.title} by {book1.author}")
print(f"{book2.title} by {book2.author}")
```

```python
class Book:

    def __init__(self,title,author):
        self.title = title
        self.author = author

    def display(self):
        print(f"{self.title} by {self.author}")

book1 = Book("Python Basic","Roy")
book2 = Book("AI Engineering","Andrew")

book1.display()
book2.display()

    
```
