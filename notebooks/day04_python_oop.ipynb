```python
class Book:

    def __init__(self,title,author):
        self.title = title
        self.author = author

book1 = Book("Python Basic","Roy")
book2 = Book("AI Engineering","Andrew")

print(f"{book1.title} by {book1.author}")
print(f"{book2.title} by {book2.author}")
```

```python
class Book:

    def __init__(self,title,author):
        self.title = title
        self.author = author

    def display(self):
        print(f"{self.title} by {self.author}")

book1 = Book("Python Basic","Roy")
book2 = Book("AI Engineering","Andrew")

book1.display()
book2.display()

    
```

```python
class Laptop:
    brand = "ASUS"
    def __init__(self,owner,ram):
        self.owner = owner
        self.ram = ram
    def display(self):
        print(f"Owner : {self.owner}\nBrand : {self.brand}\nRAM : {self.ram}")

laptop1 = Laptop("Roy", "16GB")
laptop1.display()
laptop2 = Laptop("Alice", "32GB")
laptop2.display()
```

🤔 What is a Dunder Method?

Dunder = Double Underscore.

Examples:

__init__
__str__
__len__
__getitem__
__call__

```python
class Book:

    def __init__(self,title,author):
        self.title = title
        self.author = author
    def __str__(self):
        return f"{self.title} by {self.author}"
         
book1 = Book("Python Basic","Roy")
book2 = Book("AI Engineering","Andrew")
    
print(book1)
print(book2)
```
