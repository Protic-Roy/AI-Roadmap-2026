```python
numbers = [5,2,8,1,9]
n = len(numbers)
for i in range (n-1):
    min_ind = i
    for j in range(i+1,n):
        if(numbers[j]<numbers[min_ind]):
            min_ind = j
    numbers[i],numbers[min_ind] = numbers[min_ind],numbers[i]
print(numbers)

print(f"largest number =",numbers[n-1])
print(f"Smallest number = ",numbers[0])

sum = 0
for i in range(n):
    sum+=numbers[i]
print(f"Sum of the list =",sum)


```

🎯 Mini AI Challenge

You have confidence scores:

scores = [0.91,0.42,0.88,0.73,0.31]

Create:

A list of percentages.
A list containing only scores greater than 0.5.
A dictionary:
{
0:0.91,
1:0.42,
2:0.88,
...
}

using a dictionary comprehension.

```python
scores = [0.91,0.42,0.88,0.73,0.31]
percentage ={score*100 for score in scores}
print(percentage)

high_score = {score for score in scores if(score>0.5)}
print(high_score)

dictonary_score = {index: score for index, score in enumerate(scores) }
print(dictonary_score)
```

## zip(), enumerate(), and Unpacking

Create a dictionary using zip()

```python
keys = ["model", "version"]
values = ["GPT-4", 1.0]
info = dict(zip(keys, values))

print(info)
```

Use enumerate() to print items with numbers

```python
models = ["GPT-4","Claude","Gemini"]

for index, model in enumerate(models, start=1):
    print(index, model)
```

Unpack a list into three variables

```python
numbers = [100, 200, 300]

a, b, c = numbers

print(a)  # Output: 100
print(b)  # Output: 200
print(c)  # Output: 300
```

Use extended unpacking *middle

```python
data = [10, 20, 30, 40, 50]

first, *middle, last = data

print("First:", first)    # Output: 10
print("Middle:", middle)  # Output: [20, 30, 40]
print("Last:", last)      # Output: 50
```
