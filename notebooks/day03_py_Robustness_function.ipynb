🎯 Coding Challenge

Write a calculator program that:

Takes two numbers as input.
Uses try and except to catch:
ZeroDivisionError
ValueError
Uses else to print the result if there is no error.
Uses finally to print:
Thank you for using the calculato

```python
def run_calculator():
    try: 
        num1 = float(input("Enter number 1: "))
        num2 = float(input("Enter number 2: "))
        opt = input("Enter an operation (+, -, *, /): ")
        if opt == '+':
            result = num1 + num2
        elif opt == '-':
            result = num1 - num2
        elif opt == '*':
            result = num1 * num2
        elif opt == '/':
            result = num1 / num2
        else:
            print("Invalid operation.")
            return

    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
    except ValueError:
        print("Error: Invalid input. Please enter numeric values.")     
    else:
        print(f"The result of {num1} {opt} {num2} is: {result}")
    finally:
        print("Calculator operation completed.")

run_calculator()

```
