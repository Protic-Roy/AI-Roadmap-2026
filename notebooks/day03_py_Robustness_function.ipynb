```python
def student_info(name,university,semester):
    print(f"Student : {name}\nUniversity : {university}\nSemester : {semester}")
student_info("roy","DIU",4)
```

```python
def square(number):
    a = number*number
    return a
print(square(7))
```

```python
def maximum(*numbers):
    return max(numbers)
print(maximum(3,8,2))
        
```

```python
def student_names(*names):
    print("Student LIst:\n")
    for i , name in enumerate(names,start=1):
        print(f"{i}.{name}")
student_names("roy","kill","who")
```
